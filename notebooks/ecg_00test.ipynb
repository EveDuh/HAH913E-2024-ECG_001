# 1.2.2. Display a graph of the file contents with the correct scale, labels, and format.

```python

# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Load the ECG data from the provided file
file_path = 'ECG_hz.txt'
ecg_data = np.loadtxt(file_path)

# Sampling parameters
sampling_rate = 360  # Hz
mV_per_unit = 0.5    # Conversion factor from units to millivolts

# Convert data to millivolts
ecg_data_mV = ecg_data * mV_per_unit

# Create time array based on the sampling rate
time_array = np.arange(len(ecg_data)) / sampling_rate

# Plotting the ECG signal
plt.figure(figsize=(10, 6))
plt.plot(time_array, ecg_data_mV, label='Lead I')
plt.title('ECG Signal (Lead I)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude (mV)')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
